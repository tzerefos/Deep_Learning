```python
sns.set()
main_dir = "datasets/archive/chest_xray/"
training_data = main_dir + "train/"
validation_data = main_dir + "val/"
testing_data = main_dir + "test/"

train_n = training_data+'NORMAL/'
train_p = training_data+'PNEUMONIA/'

test_n = testing_data+'NORMAL/'
test_p = testing_data+'PNEUMONIA/'

val_n = validation_data+'NORMAL/'
val_p = validation_data+'PNEUMONIA/'

train_n_size = len(os.listdir(train_n))
train_p_size = len(os.listdir(train_p))

test_n_size = len(os.listdir(test_n))
test_p_size = len(os.listdir(test_p))

val_n_size = len(os.listdir(val_n))
val_p_size = len(os.listdir(val_p))

fig,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(20,5))
ax1.bar(['Normal', 'Pneumonia'],[train_n_size, train_p_size], color=['#779ECB', '#F69592'])
ax2.bar(['Normal', 'Pneumonia'],[test_n_size, test_p_size], color=['#779ECB', '#F69592'])
ax3.bar(['Normal', 'Pneumonia'],[val_n_size, val_p_size], color=['#779ECB', '#F69592'])
ax1.title.set_text('Training dataset')
ax2.title.set_text('Testing dataset')
ax3.title.set_text('Validation dataset')
plt.show()
```
